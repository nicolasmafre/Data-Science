# Que tipo de dados o Pandas processa?

```python
import pandas as pd
```

```python
dados = {
    'Nome': [
        'José Geraldo de Freitas',
        'Silvana Martins de Toledo',
        'Nícolas Martins de Freitas',
        'Danilo Martins de Freitas'
    ],
    'Sexo': [
        'Masculino',
        'Feminino',
        'Masculino',
        'Masculino'
    ],
    'Idade': [
        65,
        55,
        39,
        37
    ],
    'Cidade': [
        'Itabera',
        'São Paulo',
        'Assis',
        'São Paulo'
    ]
}
```

```python
df = pd.DataFrame(dados)
```

```python
df['Cidade']
```

```python
idade = pd.Series([65, 55, 39, 37], name='Idade')
idade
```

## Faça algo com Dataframe ou Series

```python
# Usando o DataFrame

df['Idade'].max()
```

```python
# Usando Series

idade.max()
```
