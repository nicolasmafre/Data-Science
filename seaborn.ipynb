```python
# Importa a biblioteca Seavorn
import seaborn as sns

# Aplicando o tema principal para a aparência do gráfico
sns.set_theme()

# Carrega um exemplo da configuração de dados
tips = sns.load_dataset("tips")

# Cria uma visualização
sns.relplot(
    data=tips,
    x = "total_bill", y = "tip", col = "time",
    hue = "smoker", size = "size",
)
```
