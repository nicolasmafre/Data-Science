# Series

Uma série (Series) é uma matriz unidimensional rotulada capaz de armazenar qualquer tipo de dado (inteiros, strings, números de ponto flutuante, objetos Python, etc.). Os rótulos dos eixos são chamados coletivamente de índices. O método básico para criar uma série é chamar:

```python
# Criando e chamando uma Série
s = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'])
s
```

# Apresentação de correlações quase perfeitas com ciência de dados e aprendizagem de máquina

A amostra detalha que o crescimento da línguagem Python nos últimos 5 anos está diretamente conectado ao crescimento da importância dessas áreas.

```python
# URL do arquivo CSV com dados do Google Trends
url = 'https://tmfilho.github.io/pyestbook/data/google-trends-timeline.csv'

# Lê o CSV da URL.
# index_col=0: Usa a primeira coluna como índice (rótulos das linhas).
# parse_dates=True: Tenta interpretar o índice como datas.
trends = pd.read_csv(url, index_col=0, parse_dates=True)

# Calcula a correlação de Spearman entre as colunas.
# O método de Spearman avalia relações monotônicas (baseadas em ranking),
# útil para dados que não necessariamente têm uma relação linear perfeita.
trends.corr(method='spearman')
```

O código abaixo gera um gráfico com as quatro séries temporais de interesse. Os valores são relativos ao maior valor de interesse no período, que nesse caso equivale ao interesse por Python em setembro de 2019.


```python
plot = trends.plot()
plt.xlabel('Semana', fontsize=16)
plt.show()
```
